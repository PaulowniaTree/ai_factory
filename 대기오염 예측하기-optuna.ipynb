```python
# 위경도 값들 간의 거리 계산을 위한 함수 작성
from geopy.distance import geodesic 
def distance(lat1, lon1, lat2, lon2):
     return geodesic((lat1, lon1), (lat2, lon2)).km
```

```python
def Distance(n):
    lat  = aws_df[aws_df['Location']==aws_name[n]]['Latitude'].values[0]
    lon =  aws_df[aws_df['Location']==aws_name[n]]['Longitude'].values[0]

    meta_pm['distance'] = meta_pm.apply(lambda x: distance(x['Latitude'], x['Longitude'], lat, lon), axis=1)
    sorted_data = meta_pm.sort_values('distance')

    
    return sorted_data
```

```python
# 로그 파이프라인
log_attribs = X_train[['풍속(m/s)', '강수량(mm)','습도(%)']]

log_pipeline = Pipeline([
                        ("log_transfomer", FunctionTransformer(np.log1p)),  # 로그 변환
                        ("minmax_scaler",  MinMaxScaler()),
                        ("std_scaler", StandardScaler()) 
                              ])
```

```python
# 표준화, 정규화 파이프라인
std_attribs = X_train[['기온(°C)', '풍향(deg)']]

std_mm_pipeline = Pipeline([
                        ("minmax_scaler",  MinMaxScaler()),
                        ("std_scaler", StandardScaler()) 
                              ])
```

```python
std_attribs = ['기온(°C)', '풍향(deg)']
minmax_attribs = ['기온(°C)', '풍향(deg)']
log_attribs = ['풍속(m/s)', '강수량(mm)','습도(%)']
cat_attribs = ['Location']
full_pipeline = ColumnTransformer(
     transformers=[
                ('std_mm_pipeline', std_mm_pipeline, std_attribs),
                ('log_pipeline', log_pipeline, log_attribs),  
                ('oh_encoder', OneHotEncoder(sparse_output=False), cat_attribs), 
                #('pass', 'passthrough', ['연도', '일시','time'])
                                  ])
full_pipeline = full_pipeline.set_output(transform='pandas')
```

```python
def objective(trial):
    
    train_X, X_valid , train_y , y_valid = train_test_split(X_prepared, y_train, test_size=0.2,random_state=42)
    param = {
        'lambda': trial.suggest_float('lambda', 1e-3, 0.1),
        'alpha': trial.suggest_float('alpha', 1e-3, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.4, 1),
        'subsample': trial.suggest_float('subsample', 0.4, 1),
        'learning_rate': trial.suggest_float('learning_rate',0.0001, 0.1),
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 4,8),
        'min_child_weight': trial.suggest_int('min_child_weight', 2, 50),
        'eval_metric': 'mae'
    }
    model = xgb.XGBRegressor(**param)  
    
    model.fit(train_X,train_y, eval_set=[(X_valid, y_valid)], early_stopping_rounds=100)
    
    preds = model.predict(test_prepared)
    
    mae = mean_absolute_error(y_test, preds)
    
    return mae
          
study_xgb = optuna.create_study(
    direction='minimize',
    study_name = 'Xgboost Optuna', 
)
study_xgb.optimize(objective, n_trials=50)

print('Best trial: score {},\nparams {}'.format(study.best_trial.value,study.best_trial.params))
```
